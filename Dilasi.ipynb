```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Pastikan berada di direktori kerja
os.chdir(r"F:\KULIAH\SEMESTER 6\PENGELOLAAN CITRA DIGITAL\Jupyter Notebook\proyek filter erosi dan gaussain\Tugas")

# Fungsi untuk menampilkan gambar asli dan hasil dilasi
def tampilkan_dilasi(nama_file):
    # Baca gambar grayscale
    gambar = cv2.imread(nama_file, cv2.IMREAD_GRAYSCALE)

    # Buat kernel
    kernel = np.ones((5, 5), np.uint8)

    # Proses dilasi
    hasil_dilasi = cv2.dilate(gambar, kernel, iterations=1)

    # Simpan hasil dilasi
    nama_output = f'dilasi_{nama_file}'
    cv2.imwrite(nama_output, hasil_dilasi)

    # Tampilkan hasil
    plt.figure(figsize=(10, 4))
    plt.suptitle(f'Dilasi untuk {nama_file}', fontsize=14)

    plt.subplot(1, 2, 1)
    plt.title('Gambar Asli')
    plt.imshow(gambar, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Hasil Dilasi')
    plt.imshow(hasil_dilasi, cmap='gray')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# Jalankan untuk dua gambar
tampilkan_dilasi('1.png')
tampilkan_dilasi('2.png')
```
